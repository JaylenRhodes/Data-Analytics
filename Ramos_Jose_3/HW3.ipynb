```python
#Libraries used
import imageio
import re
import os
import matplotlib.pyplot as plt
import pandas as pd 
from collections import Counter
from nltk.corpus import stopwords
from operator import itemgetter
from pathlib import Path
from textatistic import Textatistic
from textblob import TextBlob, Word
```

Part 2:
(a) Compute word counts for each of your documents after excluding English stop words (and optionally, performing lemmatization).

```python
#Part 2.a
#Reading and pathing to each of the text files
sstory = TextBlob(Path('A Scandal In Bohemia.txt').read_text())
speech = TextBlob(Path('I Have A Dream.txt').read_text())
poem = TextBlob(Path('Thoughts On The Precipice.txt').read_text())

#Stop Words
stopwords = stopwords.words('english')

#Word Counts
items_1 = sstory.word_counts.items()
sstory_items = [item for item in items_1 if item[0] not in stopwords]
print ('A Scandal In Bohemia Word Count:')
print (len(sstory_items))
print (sstory_items) #Word Count for "A Scandal In Bohemia"
```

```python
items_2 = speech.word_counts.items()
speech_items = [item for item in items_2 if item[0] not in stopwords]
print ('I Have A Dream Word Count:')
print (len(speech_items))
print (speech_items) #Word Count for "I Have A Dream"
```

```python
items_3 = poem.word_counts.items()
poem_items = [item for item in items_3 if item[0] not in stopwords]
print ('Thoughts On The Precipice Word Count:')
print (len(poem_items))
print (poem_items) #Word Count for "Thoughts On The Precipice"
```

(b) Create and display a bar plot for each document that include word counts for the 25 most frequent words (after the above processing).

```python
#Part 2.b
#Using Item Getter to sort frequencies
#Bar plot for short story top 25
sorted_sstory = sorted(sstory_items, key=itemgetter(1), reverse=True)
sstory_top25 = sorted_sstory[:25]
df = pd.DataFrame(sstory_top25, columns=['word', 'count'])
ax = df.plot.bar(x='word', y='count', rot=75, title='A Scandal In Bohemia Top 25')
```

```python
#Bar plot for speech top 25
sorted_speech = sorted(speech_items, key=itemgetter(1), reverse=True)
speech_top25 = sorted_speech[:25]
df = pd.DataFrame(sstory_top25, columns=['word', 'count'])
ax = df.plot.bar(x='word', y='count', rot=75, title='I Have A Dream Top 25')
```

```python
#Bar plot for poem top 25
sorted_poem = sorted(poem_items, key=itemgetter(1), reverse=True)
poem_top25 = sorted_poem[:25]
df = pd.DataFrame(poem_top25, columns=['word', 'count'])
ax = df.plot.bar(x='word', y='count', rot=75, title='Thoughts On The Precipice Top 25')
```

Part 3:
(a) Use Textatistic to compute the average of the Flesch–Kincaid, Gunning Fog, SMOG, and Dale–Chall scores for each document.

```python
#Part 3.a
from pathlib import Path

#Creating dictionaries to store and display averages
sstory_path = Path('A Scandal In Bohemia.txt').read_text()
sstory_readability = Textatistic(sstory_path)
%precision 3
sstory_dict = sstory_readability.dict() #Short story dictionary
average_sstory = (sstory_dict['fleschkincaid_score'] + sstory_dict['gunningfog_score'] + sstory_dict['smog_score'] + sstory_dict['dalechall_score']) / 4
print('The average score:', average_sstory)
print(sstory_dict)
```

```python
speech_path = Path('I Have A Dream.txt').read_text()
speech_readability = Textatistic(speech_path)
%precision 3
speech_dict = speech_readability.dict() #Speech dictionary
average_speech = (speech_dict['fleschkincaid_score'] + speech_dict['gunningfog_score'] + speech_dict['smog_score'] + speech_dict['dalechall_score']) / 4
print('The average score:', average_speech)
print(speech_dict)
```

```python
poem_path = Path('Thoughts On The Precipice.txt').read_text()
poem_readability = Textatistic(poem_path)
%precision 3
poem_dict = poem_readability.dict() #Poem dictionary
average_poem = (poem_dict['fleschkincaid_score'] + poem_dict['gunningfog_score'] + poem_dict['smog_score'] + poem_dict['dalechall_score']) / 4
print('The average score:', average_poem)
print(poem_dict)
```

Part 4:
(a) Use spaCy to compute the pairwise similarity between your documents (i.e. doc. 1 to doc. 2, doc. 1 to doc. 3, doc. 2 to doc. 3).

```python
#Part 4.a
import spacy
nlp_lg = spacy.load('en_core_web_lg')
sstory_nlp = nlp_lg(Path('A Scandal In Bohemia.txt').read_text())
speech_nlp = nlp_lg(Path('I Have A Dream.txt').read_text())
poem_nlp = nlp_lg(Path('Thoughts On The Precipice.txt').read_text())

#Similarity between the Short Story and Speech
sim_sstory_speech = sstory_nlp.similarity(speech_nlp)
print(f'Similarity between "A Scandal In Bohemia" & "I Have A Dream": {sim_sstory_speech:.3}')

#Similarity between the Short Story and Poem
sim_sstory_poem = sstory_nlp.similarity(poem_nlp)
print(f'Similarity between "A Scandal In Bohemia" & "Thoughts On The Precipice": {sim_sstory_poem:.3}')

#Similarity between the Speech and Short Story
sim_speech_sstory = speech_nlp.similarity(sstory_nlp)
print(f'Similarity between "I Have A Dream" & "A Scandal In Bohemia": {sim_speech_sstory:.3}')

#Similarity between the Speech and Poem
sim_speech_poem = speech_nlp.similarity(poem_nlp)
print(f'Similarity between "I Have A Dream" & "Thoughts On The Precipice": {sim_speech_poem:.3}')

#Similarity between the Poem and Short Story
sim_poem_sstory = poem_nlp.similarity(sstory_nlp)
print(f'Similarity between "Thoughts On The Precipice" & "A Scandal In Bohemia": {sim_poem_sstory:.3}')    

#Similarity between the Poem and Speech
sim_poem_speech = poem_nlp.similarity(speech_nlp)
print(f'Similarity between "Thoughts On The Precipice" & "I Have A Dream": {sim_poem_speech:.3}')
```

(b) Produce a bar plot for each document that includes the count for the 20 most common named entities (by name).

```python
#Part 5.b
#Function to plot entities of files
def bar_graph(entities, title, count=20):
  df = pd.DataFrame(entities[:count], columns=['word', 'count'])
  df.plot.bar(x='word', y='count', legend=False, title=title)
  plt.gcf().tight_layout()
```

```python
#Entities bar graph for A Scandal In Bohemia
sstory_top20 = sorted(dict(Counter(entity[0] for entity in sstory_entities)).items(), key=itemgetter(1), reverse=True)
bar_graph(sstory_top20, 'A Scandal In Bohemia Top 20')
```

```python
#Entities bar graph for I Have A Dream
speech_top20 = sorted(dict(Counter(entity[0] for entity in speech_entities)).items(), key=itemgetter(1), reverse=True)
bar_graph(speech_top20, 'I Have A Dream Top 20')
```

```python
#Entities bar graph for Thoughts On The Precipice
poem_top20 = sorted(dict(Counter(entity[0] for entity in poem_entities)).items(), key=itemgetter(1), reverse=True)
bar_graph(poem_top20, 'Thoughts On The Precipice Top 20')
```

(c) Produce a second bar plot per document based on the counts of every named entity type (PERSON, ORG, etc.)

```python
#Part 5.c
#Function to plot entity groups of files
def bar_graph_groups(entities, title):
  df = pd.DataFrame(entities, columns=['entity_type', 'count'])
  df.plot.bar(x='entity_type', y='count', legend=False, title=title)
  plt.gcf().tight_layout()
```

```python
#Entities bar graph for Thoughts On The Precipice
sstory_groups = sorted(dict(Counter(entity[1] for entity in sstory_entities)).items(), key=itemgetter(1), reverse=True)
bar_graph_groups(sstory_groups, 'A Scandal In Bohemia Entity Groups')
```

```python
#Entities bar graph for Thoughts On The Precipice
speech_groups = sorted(dict(Counter(entity[1] for entity in poem_entities)).items(), key=itemgetter(1), reverse=True)
bar_graph_groups(speech_groups, 'I Have A Dream Entity Groups')
```

```python
#Entities bar graph for Thoughts On The Precipice
poem_groups = sorted(dict(Counter(entity[1] for entity in poem_entities)).items(), key=itemgetter(1), reverse=True)
bar_graph_groups(poem_top20, 'Thoughts On The Precipice Entity Groups')
```
