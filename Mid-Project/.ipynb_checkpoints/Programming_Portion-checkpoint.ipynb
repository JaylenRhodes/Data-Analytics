```python
#Importing the libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
```

```python
#calling the csv file as fname
fname = os.path.join("MetObjects_Subset.csv")
#Reading the file through the pandas library
df_init = pd.read_csv(fname)
#Printing the file
df_init
```

```python
percentage = 50.0 #This is the percentage of NaN that we want to erase
drop_count = int(((100-percentage)/100)*df_init.shape[0] + 1) #Formula that takes in the percentage variable value and then 
df_prob2 = df_init.dropna(axis=1, thresh=drop_count)
df_prob2
```

```python
#Getting the 10 most frequent names for the names in the ObjectName column
n = 10 #n is the number of columns/items
list3 = df_prob2['Object Name'].value_counts().index.tolist()[:n] #Creating a list that can filter through the row 'Object Names' and indexes the top n values
df_prob3 = pd.DataFrame(list3) #Converting the previously gathered list to be a dataframe
df_prob3
```

```python
#Creating the dictionary, setting keys and initializing values
dict_prob4={'United States': 0, 'Canada': 0, 'Mexico': 0, 'Other': 0}
#Finding the percentages
#Here I had to set the data-frame as 2 not 3 because my 3 is single column without 'Country'
percent = ((df_prob2['Country'].value_counts(dropna=False)/((df_prob2['Country'].value_counts(dropna=False).sum())))*100)
#Loop to iterate through and set the percent for each key
if(percent.get(key='United States') != None):
    dict_prob4['United States']=percent.get(key='United States')
if (percent.get(key='Canada') != None):
    dict_prob4['Canada']=percent.get(key='Canada')
if (percent.get(key='Mexico') != None):
    dict_prob4['Mexico']=percent.get(key='Mexico')
#Setting the percentage for others
dict_prob4['Other'] = 100 - (dict_prob4['United States'] + dict_prob4['Canada'] + dict_prob4['Mexico'])
dict_prob4
```

```python
#Creating a dictionary
dict_prob5=dict()
#Setting the restrictions for the dates
#Had to use data-frame 2 again as 3 is useless
dict_prob5['1900']=df_prob2[(1900 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1910)].value_counts(dropna=False).sum()
dict_prob5['1910']=df_prob2[(1910 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1920)].value_counts(dropna=False).sum()
dict_prob5['1920']=df_prob2[(1920 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1930)].value_counts(dropna=False).sum()
dict_prob5['1930']=df_prob2[(1930 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1940)].value_counts(dropna=False).sum()
dict_prob5['1940']=df_prob2[(1940 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1950)].value_counts(dropna=False).sum()
dict_prob5['1950']=df_prob2[(1950 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1960)].value_counts(dropna=False).sum()
dict_prob5['1960']=df_prob2[(1960 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1970)].value_counts(dropna=False).sum()
dict_prob5['1970']=df_prob2[(1970 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1980)].value_counts(dropna=False).sum()
dict_prob5['1980']=df_prob2[(1980 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 1990)].value_counts(dropna=False).sum()
dict_prob5['1990']=df_prob2[(1990 <= df_prob2['Object End Date']) & (df_prob2['Object End Date'] < 2000)].value_counts(dropna=False).sum()

dict_prob5
```

```python
#Converting the dictionary into a matplot graph
plt.bar(dict_prob4.keys(), dict_prob4.values(), color='b')
```
